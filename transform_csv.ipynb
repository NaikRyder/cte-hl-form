```python
import pandas as pd
from google.colab import drive
import json
```

```python
################# Input Layer #################

# Read Excel workbooks from gdrive
# Survey raw data
file_id="1Zw_vLtvL4jQmWpOGC4tYej11Zet3nofK"
transformed_results=pd.ExcelFile(f"https://docs.google.com/spreadsheets/d/{file_id}/export?format=xlsx")
```

```python
# Get number of sheets in the workbook and parse into a dataframe array
df = {}
sheet_names = transformed_results.sheet_names
sheet_count=len(sheet_names)-1
for i in range(0,sheet_count):
  df[i] = transformed_results.parse(i)
  df[i] = df[i].drop(['First Name', 'Last Name', 'Email Address', 'Timestamp'], axis =1)
```

```python
t_df = pd.DataFrame()
for i in range(1,sheet_count):
  t_df = pd.merge(df[0],df[i], how='left', on='Student ID')
```

```python
t_df
```

```python
# Convert dataframe into a .csv file and save it to gdrive
drive.mount('/content/drive')
path = '/content/drive/My Drive/Survey Results/survey_data/consolidated_results.xlsx'
with pd.ExcelWriter(path) as writer:
  t_df.to_excel(writer,index=False)
```

```python
t_df_json = json.loads(t_df.to_json(orient = 'records'))
json_data = []
indent = 3

for item in t_df_json:
    student_id = item.pop("Student ID")
    item["Student ID"] = student_id
    new_item = {"Student ID": student_id, "Response": item}
    json_data.append(new_item)

json_data = json.dumps(json_data, indent=indent)
```

```python
# json_file = "https://drive.google.com/file/d/1l9ULqS9oOxJ7h2909Cs5eDN7S_HoFAVf"
with open("/content/drive/MyDrive/TEILab-Work/GD-ITEST-CTE-Team-Folder/CTE_2022_2023/Hearne-Summer2023-Project/Data/Survey Results/survey_data/survey_results.json", "w") as json_file:
    json.dump(json_data, json_file, indent=indent)
```
